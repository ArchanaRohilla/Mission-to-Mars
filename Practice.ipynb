```python
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
# Windows users
executable_path = {'executable_path':'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#visit the quotes to scrape site.
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
#parse the html
html = browser.html
soup =  BeautifulSoup(html, 'html.parser')
```

```python
#scrape the title
title = soup.find('h2').text
title
```

```python
#scrape the top ten tags
tag_box = soup.find('div', class_='tags-box')

#tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
#this section of code will scrape the quotes on the first page, click the "Next" button,
#then scrape more quotes and so on (five pages worth of quotes).
for x in range(1,6):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    quotes = soup.find_all('span', class_= 'text')
    for quote in quotes:
        print('page:', x, '-------------')
        print(quote.text)
    browser.click_link_by_partial_text('Next') 
    
```

```python
#skill-drill
url = 'http://books.toscrape.com/'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
my_list = soup.find('div', class_='side_categories')
names = my_list.find_all('a')
 
for name in names:
    word = name.text
    print(word)
```
